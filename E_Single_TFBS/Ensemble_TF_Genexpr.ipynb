## Correlation between TF Expression and mean TF-Targetgene Expression

```python
mean_log2_expr_per_tissue = pd.read_csv("images/CSV_log2_mean_GenExr_diffrent_tissue_all_TFBS_ROBUST_extend_with_traget_gene_count.csv").set_index("tf")
```

```python
mean_log2_expr_per_tissue
```

```python
def get_pearson_for_TF_and_Target_Gen(target_expr, tf_expr):
    r_for_tfs = []
    p_for_tfs = []
    tf_lst = []
    for tf in tf_expr.index:
        mean_target_expr = target_expr.loc[tf]
        single_tf_expr = tf_expr.loc[tf]
        
         # REMOVE NANS for both Arrays
        mask = ~np.isnan(single_tf_expr)
        mean_target_expr = mean_target_expr[mask]
        single_tf_expr = single_tf_expr[mask]
        
        if len(single_tf_expr) < 2:
            print(f"{tf} could not be calculated, because there are too many NAN values. The arrays must have at least length 2.")
        else:
            r, p = pearsonr(mean_target_expr, single_tf_expr)
            r_for_tfs.append(r)
            p_for_tfs.append(p)
            tf_lst.append(tf)
    df_corr_tf_expr_mean_expr =  pd.DataFrame({"TFs":tf_lst, "Pearson_r":r_for_tfs, "Pearson_p":p_for_tfs})
    return df_corr_tf_expr_mean_expr
```

```python
df_corr_tf_expr_mean_expr = get_pearson_for_TF_and_Target_Gen(mean_log2_expr_per_tissue, expr_for_tf_df_log)
```

```python
target_tf_expression_corr = df_corr_tf_expr_mean_expr[df_corr_tf_expr_mean_expr["Pearson_p"] < 0.05]
target_tf_expression_corr
```

```python
# Which TFs have to high p-vlaue ?? 
np.array(sorted(df_corr_tf_expr_mean_expr[df_corr_tf_expr_mean_expr["Pearson_p"] > 0.05].TFs))
```

```python
fig, ax = plt.subplots(figsize=(10,30))

norm = plt.Normalize(target_tf_expression_corr["Pearson_p"].min(), target_tf_expression_corr["Pearson_p"].max())
sm = plt.cm.ScalarMappable(cmap="rocket_r", norm=norm)

sns.barplot(target_tf_expression_corr, x="Pearson_r", y="TFs",hue="Pearson_p", width=0.5, palette="rocket_r")


ax.set_ylabel('TFBS')

ax.xaxis.tick_top()
ax.set_xlabel('Pearson Correlation')
ax.xaxis.set_label_position("top")

ax.xaxis.grid(True)
ax.get_legend().remove()

plt.subplots_adjust(top=0.9)
cbar = ax.figure.colorbar(sm, ax=ax, orientation= "horizontal", location="top", )
cbar.set_label("-log10(p-value)")

#fig.savefig("/sybig/projects/GeneRegulation/data/jme/Bachelorarbeit/data/Single_TFBS/Boxplot/Pearson_corr_all_great_p.png")
```
