## Correlation between TF Expression and mean TF-Targetgene Expression

```python
def flatten_tissues_expand_df(tfbs_df):
    tfbs_part_repeat = tfbs_df.iloc[:,:9]
    tfbs_part_flatten = tfbs_df.iloc[:,9:]
    
    repeat_df =  pd.DataFrame(np.repeat(tfbs_part_repeat.values, tfbs_part_flatten.shape[1], axis=0))
    flatten_arr = tfbs_part_flatten.to_numpy().flatten()

    tissue_repeat = np.tile(np.array(tfbs_part_flatten.columns), len(tfbs_part_repeat))
    
    repeat_df.columns = tfbs_part_repeat.columns
    repeat_df["GeneExpr"] = flatten_arr
    repeat_df["Tissue"] = tissue_repeat
    

    return repeat_df
```

```python
flat_df = flatten_tissues_expand_df(df_combined)
pivot_df = flat_df.pivot_table(index="tf", columns="Tissue", values="GeneExpr", aggfunc="mean")
pivot_df
```

```python
tf_expr = expr_for_tf_df.copy()
tf_expr.loc["chrf28811"] = np.full(54,np.nan)
tf_expr.loc["T"] = np.full(54,np.nan)
tf_expr.loc["ARNTL"] = np.full(54,np.nan)
tf_expr.loc["SOX10"] = np.full(54,np.nan)

tf_expr_sort = tf_expr.sort_index()
tf_expr_sort
```

```python
r_for_tfs = []
p_for_tfs = []
tf_lst = []
for tf in tf_expr_sort.index:
    mean_target_expr = pivot_df.loc[tf]
    single_tf_expr = tf_expr_sort.loc[tf]
    if np.any(np.isnan(single_tf_expr)):
        print(f"{tf} could not be calculated, due to NaN values.")
    else:
        r, p = pearsonr(mean_target_expr, single_tf_expr)
        r_for_tfs.append(r)
        p_for_tfs.append(p)
        tf_lst.append(tf)
df_corr_tf_expr_mean_expr =  pd.DataFrame({"TFs":tf_lst, "Pearson_r":r_for_tfs, "Pearson_p":p_for_tfs})
```

```python
df_corr_tf_expr_mean_expr
```

```python
df_corr_tf_expr_mean_expr_filter_p = df_corr_tf_expr_mean_expr[df_corr_tf_expr_mean_expr["Pearson_p"] < 0.05].copy()
df_corr_tf_expr_mean_expr_filter_p["Pearson_p"] = -np.log10(df_corr_tf_expr_mean_expr_filter_p["Pearson_p"])
```

```python
fig, ax = plt.subplots(figsize=(10,30))

norm = plt.Normalize(df_corr_tf_expr_mean_expr["Pearson_p"].min(), df_corr_tf_expr_mean_expr["Pearson_p"].max())
sm = plt.cm.ScalarMappable(cmap="rocket_r", norm=norm)

sns.barplot(df_corr_tf_expr_mean_expr, x="Pearson_r", y="TFs",hue="Pearson_p", width=0.5, palette="rocket_r")


ax.set_ylabel('TFBS')

ax.xaxis.tick_top()
ax.set_xlabel('Pearson Correlation')
ax.xaxis.set_label_position("top")

ax.xaxis.grid(True)
ax.get_legend().remove()

plt.subplots_adjust(top=0.9)
cbar = ax.figure.colorbar(sm, ax=ax, orientation= "horizontal", location="top", )
cbar.set_label("-log10(p-value)")

#fig.savefig("/sybig/projects/GeneRegulation/data/jme/Bachelorarbeit/data/Single_TFBS/Boxplot/Pearson_corr_all_great_p.png")
```
